# Example-3
* 1-50 arasinda rastgele 5 sayi uretsin,
* show butonu ile ekrana yazdirsin,
* transparent butonu ile ekran saydamlastirsin
* default butonu ile ayarlar default hale gelsin
* full screen butonu ie tam ekran yap,
* min screen butonu ile kucuk ekran yap

```python
from random import sample
import tkinter as tk

form = tk.Tk()
form.title('Example-3')
form.geometry('420x250+500+250')
form.state('normal')

def show_numbers():
    label.config(text= str(sorted(sample(range(1,51),5))))
    
def transparent_screen():
    form.wm_attributes('-alpha',0.6)    
    
def default_screen():
    form.wm_attributes('-alpha',1)  
    
def full_screen():
    form.state('zoomed')
    
def min_screen():
    form.state('normal')
    form.geometry('420x250+500+250')
    
label= tk.Label(form, fg='blue', bg='grey',font= 'ariel 17 bold')
label.pack()  

show_button = tk.Button(form, text= 'show', fg='black', bg='red', font= 'ariel 12 bold',command = show_numbers)
show_button.pack(side=tk.LEFT)

transparent_button = tk.Button(form, text= 'transparent', fg='white', bg='limegreen', font= 'ariel 12 bold',command = transparent_screen)
transparent_button.pack(side=tk.LEFT)

default_button = tk.Button(form, text= 'default', fg='black', bg='darkgreen', font= 'ariel 12 bold',command = default_screen)
default_button.pack(side=tk.LEFT)

min_button = tk.Button(form, text= 'min screen', fg='grey', bg='maroon', font= 'ariel 12 bold',command = min_screen)
min_button.pack(side=tk.LEFT)

full_button = tk.Button(form, text= 'full screen', fg='grey', bg='navy', font= 'ariel 12 bold',command = full_screen)
full_button.pack(side=tk.LEFT)

form.mainloop()
```

```python
import tkinter as tk

form = tk.Tk()
form.title('Tkinter Examples') 

form.geometry('500x250+500+150')
form.state('normal')

entry = tk.Entry(form)
entry.pack()

def get_data():
    label['text'] = entry.get() # default text degere sahip label in text i degistirilebilir.

def remove():
    entry.delete(0,'end')
    
label= tk.Label(form, text='Lorem ipsum dolor sit amet', fg='blue',font= 'ariel 17 bold') 
label.pack()  

get_button = tk.Button(form, text= 'GET ENTRY', fg='black', bg='green', font= 'ariel 18 bold',command = get_data)
get_button.pack(anchor='s',
                fill=tk.X,
                padx=10,pady=20,
                ipady=10)

remove_button = tk.Button(form, text= 'REMOVE ENTRY', fg='black', bg='red', font= 'ariel 18 bold',command = remove)
remove_button.pack(anchor='s',
                    fill=tk.X,
                    ipady=10) # butonun x ve y boyutlarini belirler.

form.mainloop()
```

# Example-6
* User Name ve Password u alip ekrana yazdiran,
* remove butonu ile formlari temizleyen app yapiniz.

```python
import tkinter as tk

form = tk.Tk()
form.title('Tkinter Examples') 

form.geometry('500x400+500+150')
form.state('normal')

user_label= tk.Label(form, text='User Name', fg='black',font= 'ariel 17 bold') 
user_label.pack()  

user_name = tk.Entry(form, fg='black',font= 'ariel 17 normal') 
user_name.pack()

password_label= tk.Label(form, text='Password', fg='black',font= 'ariel 17 bold') 
password_label.pack()

password = tk.Entry(form, show='*',fg='black',font= 'ariel 17 normal') 
password.pack()

def get_data():
    label['text'] = f"Username: {user_name.get()}\nPassword: {password.get()}" 
    
def remove():
    user_name.delete(0,'end')
    password.delete(0,'end')
    
label= tk.Label(form, text='Lorem ipsum dolor sit amet', fg='blue',font= 'ariel 17 bold') 
label.pack() 

button_1 = tk.Button(form, text= 'GET DATA', fg='black', bg='green', font= 'ariel 18 bold',command = get_data)
button_1.pack(anchor='s',
            fill=tk.X,
            padx=10,pady=20,
            ipady=10)

button_2 = tk.Button(form, text= 'REMOVE ENTRY', fg='black', bg='red', font= 'ariel 18 bold',command = remove)
button_2.pack(anchor='s',
            fill=tk.X,
            padx=10,pady=20,
            ipady=10) # butonun x ve y boyutlarini belirler.

form.mainloop()
```

# Check Button

```python
import tkinter as tk

form = tk.Tk()
form.title('Check Button') 

form.geometry('500x250+500+150') # GUI nin default x,y boyutlari ve desktopta acilacagi konum koordinatlari belirler
form.state('normal') # GUI, desktopta acilir

x=tk.IntVar()
x.set(0)

def check():
    if x.get()==1:
        label= tk.Label(form, text= 'validate', fg='green', bg='black') # fg:frontground, bg:background
        label.pack()
    else:
        label= tk.Label(form, text= 'unvalidate', fg='red', bg='black') # fg:frontground, bg:background
        label.pack()

check_button = tk.Checkbutton(form, text= 'check box', fg='black', bg='blue', font= 'ariel 12 bold',variable=x, command = check)
check_button.pack(side=tk.LEFT)

form.mainloop() # GUI yi acik tutar.
```

# Radio Button

```python
import tkinter as tk

form = tk.Tk()
form.title('Radio Button') 

form.geometry('500x250+500+150') # GUI nin default x,y boyutlari ve desktopta acilacagi konum koordinatlari belirler
form.state('normal') # GUI, desktopta acilir

x=tk.StringVar()

def check():
    if x.get()=='1':
        label= tk.Label(form, text= 'radio_1', fg='red', bg='black')
        label.pack()
    elif x.get()=='2':
        label= tk.Label(form, text= 'radio_2', fg='green', bg='black')
        label.pack()
    else:
        label= tk.Label(form, text= 'radio_1 and radio_2', fg='blue', bg='black')
        label.pack()
       

button = tk.Button(form, text= 'click', fg='black', bg='blue', font= 'ariel 12 bold',activebackground='green', command = check)
button.pack(side=tk.LEFT)        

radio_1 = tk.Radiobutton(form, text= 'radio_1',activebackground='red',value=1, variable=x)
radio_1.pack()

radio_2 = tk.Radiobutton(form, text= 'radio_2',activebackground='green',value=2, variable=x)
radio_2.pack()

form.mainloop()
```

# Message Box

```python
import tkinter as tk
from tkinter import messagebox

form = tk.Tk()
form.title('Message Box') 

form.geometry('500x250+500+150') 
form.state('normal')

def message():
    messagebox.showinfo(title='message1', message='message 1') # tamam
    messagebox.askretrycancel(title='message2', message='message 2') # yeniden dene, iptal
    messagebox.askyesno(title='message3', message='message 3') # evet, hayir
#     messagebox.askquestion(title='message4', message='message 4') # askyesno ile ayni
    messagebox.askyesnocancel(title='message5', message='message 5') # evet, hayir, iptal

button = tk.Checkbutton(form, text= 'message box', command = message).pack()

form.mainloop()
```
